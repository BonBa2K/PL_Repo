```python
import pandas as pd

Yu_Gi_Oh_card_data= pd.read_csv("Yu_Gi_Oh_card_data.csv")
Yu_Gi_Oh_card_data.head(4)
```

```python
#
Winged_Beast = Yu_Gi_Oh_card_data[Yu_Gi_Oh_card_data['Race'].str.contains("Winged Beast")]
Winged_Beast.head(4)
```

```python
#4
Lv_4 = Yu_Gi_Oh_card_data[Yu_Gi_Oh_card_data['Level'] <= 4]
Lv_4.head(4)
```

```python
#  <=4 
Target_intersection=set(list(Winged_Beast['Name'])).intersection(set(list(Lv_4['Name'])))
Yu_Gi_Oh_card_data.loc[Yu_Gi_Oh_card_data.Name.isin(Target_intersection)].head(4)
```

```python
#  >4
Target_difference=set(list(Winged_Beast['Name'])).difference(set(list(Lv_4['Name'])))
Yu_Gi_Oh_card_data.loc[Yu_Gi_Oh_card_data.Name.isin(Target_difference)].head(4)
```

```python
#  <=4 
Target_union=set(list(Winged_Beast['Name'])).union(set(list(Lv_4['Name'])))
Yu_Gi_Oh_card_data.loc[Yu_Gi_Oh_card_data.Name.isin(Target_union)].head(4)
```

```python
#   >4 
Target_symmetric_difference=set(list(Winged_Beast['Name'])).symmetric_difference(set(list(Lv_4['Name'])))
Yu_Gi_Oh_card_data.loc[Yu_Gi_Oh_card_data.Name.isin(Target_symmetric_difference)].head(4)
```
